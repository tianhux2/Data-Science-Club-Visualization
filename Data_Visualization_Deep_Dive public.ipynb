```python
import pandas as pd
from matplotlib import colormaps
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("building_inventory.csv", na_values={'Square Footage': 0, 
                                                    'Year Constructed':0,
                                                    'Year Acquired':0}) 
```

```python
df.head()
```

```python
df.columns
```

### Total Floors v Building Status

```python
plt.scatter(x=df["Total Floors"], y=df["Bldg Status"])
plt.xlabel("Total Floors")
plt.ylabel("Bldg Status")
plt.show()
```

### bar chart

```python
plt.bar(df["Year Constructed"], df["Total Floors"])
plt.xlabel("Year Constructed")
plt.ylabel("Total Floors")
plt.show()
```
